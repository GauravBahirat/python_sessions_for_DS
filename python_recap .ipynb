```python
# write a program 

# ask the user enter a basic salary 
# ask the user enter DA amount 
# ask the user enter HRA amount 
# calculate total pay = basic + DA+ HRA 
```

```python
basic_salary=eval(input('Enter the basic salary:'))
DA=eval(input('Enter DA amount:'))
HRA=eval(input('Enter HRA amount:'))
total_pay=basic_salary+DA+HRA
print('Your total Pay is : {}'.format(total_pay))
```

Q:4> ask the user enter total bill in dollars,
